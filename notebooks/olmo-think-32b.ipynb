```python
from openai import OpenAI
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
OLMO_THINK_API = os.getenv('OLMO_THINK_API')
OLMO_THINK_BASE_URL = os.getenv('OLMO_THINK_BASE_URL')
```

```python
client = OpenAI(
  base_url=OLMO_THINK_BASE_URL,
  api_key=OLMO_THINK_API,
)
```

```python
response = client.chat.completions.create(
  model="allenai/olmo-3-32b-think:free",
  messages=[
          {
            "role": "user",
            "content": "generate 10 placement question in google"
          }
        ],
  extra_body={"reasoning": {"enabled": True}}
)
```

```python
response = response.choices[0].message.content
response
```
